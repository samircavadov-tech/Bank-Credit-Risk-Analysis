```python
import pandas as pd

# Datanı oxumaq
df = pd.read_csv(r"C:\Users\cavad\OneDrive\Desktop\məşq\credit_risk_dataset.csv")

# Hər bir sütundakı cəmi null (boş) dəyərlərin sayını tapmaq
print("İlkin boş dəyərlər sayı:\n", df.isnull().sum())

# NULL dəyərləri olan sütunların statiskasına baxmaq
print(df[['person_emp_length', 'loan_int_rate']].describe())

#  İş stajı boş olanları median (orta göstərici) ilə doldururuq
df['person_emp_length'] = df['person_emp_length'].fillna(df['person_emp_length'].median()).round(2)

#  Kredit faiz dərəcəsi boş olanları ortalama (mean) faizlə doldururuq
df['loan_int_rate'] = df['loan_int_rate'].fillna(df['loan_int_rate'].mean()).round(2)

# Yoxlamaq üçün yenidən null dəyərlərin sayına baxırıq


# Datada duplikat sətr sayı
print("İlkin duplikat sətr sayı:", df.duplicated().sum())

# Duplikatların silinməsi
df=df.drop_duplicates()

# Təmizlik işindən sonra nəticə
print("NULL dəyərlər:", df.isnull().sum())
print("Duplikat sətrlər:" , df.duplicated().sum())

# Təmizlənmiş dataseti yaddaşda saxlamaq
df.to_csv("cleaned_credit_risk.csv", index=False)
print("Tərtəmiz data 'cleaned_credit_risk.csv' olaraq qovluğa yazıldı!")
```
